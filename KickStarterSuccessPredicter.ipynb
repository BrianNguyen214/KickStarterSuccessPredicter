```python
def getValidDFOfCategory(theBackAvg, theGoalAvg, thePledgeAvg, theBackStd, theGoalStd, thePledgeStd, theCategoryDF):
    validCategoryAndStats = {}
    validID = []
    validName = []
    validCategory = []
    validMainCategory = []
    validCurrency = []
    validDeadline = []
    validGoal = []
    validLaunched = []
    validPledged = []
    validState = []
    validBackers = []
    validCountry = []
    validUSDPledged = []
    validUSDPledgedReal = []
    validUSDGoalReal = []
    #print(theBackAvg)
    print(len(theCategoryDF))
    for i, row in theCategoryDF.iterrows():
        #print(row['name'])
        if row['backers'] > theBackAvg - 2*theBackStd and row['backers'] < theBackAvg + 2*theBackStd and row['goal'] > theGoalAvg - 2*theGoalStd and row['goal'] < theGoalAvg + 2*theGoalStd and row['pledged'] > thePledgeAvg - 2*thePledgeStd and row['pledged'] < thePledgeAvg + 2*thePledgeStd:
            validID.append(row['ID'])
            validName.append(row['name'])
            validCategory.append(row['category'])
            validMainCategory.append(row['main_category'])
            validCurrency.append(row['currency'])
            validDeadline.append(row['deadline'])
            validGoal.append(row['goal'])
            validLaunched.append(row['launched'])
            validPledged.append(row['pledged'])
            validState.append(row['state'])
            validBackers.append(row['backers'])
            validCountry.append(row['country'])
            validUSDPledged.append(row['usd pledged'])
            validUSDPledgedReal.append(row['usd_pledged_real'])
            validUSDGoalReal.append(row['usd_goal_real'])
    validCategoryAndStats = {
        'ID': validID,
        'name': validName,
        'category': validCategory,
        'main_category': validMainCategory,
        'currency': validCurrency,
        'deadline': validDeadline,
        'goal': validGoal,
        'launched': validLaunched,
        'pledged': validPledged,
        'state': validState,
        'backers': validBackers,
        'country': validCountry,
        'usd pledged': validUSDPledged,
        'usd_pledged_real': validUSDPledgedReal,
        'usd_goal_real': validUSDGoalReal
    }
    
    return validCategoryAndStats

listOfDfValidMainCategoriesData = [0] * 15
for i in range(len(listOfDfMainCategories)):
    valMCandData = getValidDFOfCategory(backAvgs[i], goalAvgs[i], pledgeAvgs[i], backSTDs[i], goalSTDs[i], pledgeSTDs[i], listOfDfMainCategories[i])
    listOfDfValidMainCategoriesData[i] = pd.DataFrame(valMCandData)
#print(listOfDfValidMainCategoriesData[0])
```
